```python
import sys
if ".." not in sys.path:
    sys.path.insert(0, "..")
    
from src.utils import *
from src.plots import *
import torch.optim as optim
```

```python
embs_wavlm = torch.load('../embeddings/audio/raw_wavlm.pt')
embs_video = torch.load('../embeddings/video/raw.pt')
fused_embs = concat_embs(embs_wavlm, embs_video)
```

```python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = nn.CrossEntropyLoss()
num_epochs = 50
train_loader, test_loader = train_test_dataloader(fused_embs)
```

```python
PARAMS = {'criterion': criterion,
        'num_epochs': num_epochs,
        'train_loader': train_loader,
        'test_loader': test_loader,
        'device': DEVICE,
        'patience': 10}
```

### Learning Rate

```python
learning_rates = [0.1, 0.01, 0.001, 0.0001]
losses = []
accs = []

for lr in learning_rates:
    model = MLP(input_dim = 1536, hidden_dims = [512], output_dim = 6, dropout = 0.2).to(DEVICE)
    optimizer = optim.Adam(model.parameters(), lr = lr)
    params = PARAMS.copy()
    params['model'] = model
    params['optimizer'] = optimizer
    epochs, train_losses, val_losses, train_accs, val_accs, train_precs, val_precs, train_recs, val_recs, best_epoch, cm = train_model(**params)
    plot_metrics(train_losses, val_losses, train_accs, val_accs, train_precs, val_precs, train_recs, val_recs)
    losses.append(val_losses[best_epoch])
    accs.append(val_accs[best_epoch])


df = pd.DataFrame({'learning rate': learning_rates, 'loss': losses, 'accuracy': accs})
```

### Dropout

```python
dropouts = [0, 0.1, 0.2, 0.3, 0.5, 0.8]
losses = []
accs = []

for drop in dropouts:
    model = MLP(input_dim = 1536, hidden_dims = [512], output_dim = 6, dropout = drop).to(DEVICE)
    optimizer = optim.Adam(model.parameters(), lr = 0.001)
    params = PARAMS.copy()
    params['model'] = model
    params['optimizer'] = optimizer
    epochs, train_losses, val_losses, train_accs, val_accs, train_precs, val_precs, train_recs, val_recs, best_epoch, cm = train_model(**params)
    losses.append(val_losses[best_epoch])
    accs.append(val_accs[best_epoch])
    print('\n\n')


df = pd.DataFrame({'dropout': dropouts, 'loss': losses, 'accuracy': accs})
```

```python
df
```

### Hidden layers

```python
layers = [[256], [512], [1024], [512, 256], [1024, 512], [1024, 512, 256]]
losses = []
accs = []

for hidden in layers:
    model = MLP(input_dim = 1536, hidden_dims = hidden, output_dim = 6, dropout = 0.2).to(DEVICE)
    optimizer = optim.Adam(model.parameters(), lr = 0.001)
    params = PARAMS.copy()
    params['model'] = model
    params['optimizer'] = optimizer
    epochs, train_losses, val_losses, train_accs, val_accs, train_precs, val_precs, train_recs, val_recs, best_epoch, cm = train_model(**params)
    losses.append(val_losses[best_epoch])
    accs.append(val_accs[best_epoch])
    print('\n\n')
    

df = pd.DataFrame({'layers arch': layers, 'loss': losses, 'accuracy': accs})
```
